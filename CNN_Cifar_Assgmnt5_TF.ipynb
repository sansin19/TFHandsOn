```
print(tf.__version__)
```

```
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()
print("train_images shape:", train_images.shape)
print("test_images shape:", test_images.shape)
print("train_labels shape:", train_labels.shape)
print("test_labels shape:", test_labels.shape)
```

```
train_images = train_images / 255
test_images = test_images / 255
```

```
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(64,(3,3),activation='relu',input_shape=(32,32,3)))
model.add(tf.keras.layers.MaxPooling2D(pool_size = (2,2)))
#keras.layers.MaxPooling2D(pool_size=(2, 2), strides=None, padding='valid', data_format=None)
model.add(tf.keras.layers.Conv2D(32,(3,3),activation = 'relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size = (2,2)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64,activation='relu'))
model.add(tf.keras.layers.Dense(10, activation ='softmax'))
model.summary()

ko
```

```
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
#model.fit(train_images, train_labels, epochs=5, batch_size=128, shuffle=True)--72.64%
#model.fit(train_images, train_labels, epochs=5) --train acc = 77.90 %
model.fit(train_images, train_labels, epochs=10, batch_size=128, shuffle=True) #train using pooling with 10 epoch accuracy = 70.24%
```

```
loss, accuracy = model.evaluate(test_images, test_labels)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
